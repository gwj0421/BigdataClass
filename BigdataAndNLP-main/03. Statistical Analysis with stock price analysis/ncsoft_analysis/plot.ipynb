```python
resource = '/gdrive/MyDrive/College/4_1/빅데이터분석/3주차/testcsv(수정).csv'
df = pd.read_csv(resource)
```

```python
df = df.astype({'date':'str'})
df['date']
```

```python
fig = plt.figure(figsize=(20, 6))
ax = fig.add_subplot(111)
# Y축의 스케일이 다른 그래프를 ax1과 동일한 영역에 생성

lns1=ax.plot(df['date'],df['ncsoft'],ls='--',marker='o',color='black',label='NCSOFT search')
lns2=ax.plot(df['date'],df['nc_stock'],ls='--',marker='o',color='limegreen',label='nc_stock search')
lns3=ax.plot(df['date'],df['linege'],ls='--',marker='o',color='violet',label='linege search')
lns4=ax.plot(df['date'],df['nc'],ls='--',marker='o',color='dodgerblue',label='nc search')

plt.xticks(rotation=45)
ax2 = ax.twinx()
lns5=ax2.plot(df['date'],df['stock_price'],ls='-',marker='o',color='red',label='stock price')

# added these three lines
lns = lns1+lns2+lns3+lns4+lns5
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc=0)
ax.grid()

plt.show()
```

```python
df['se_volatility']=0.
len(df)
for i in range(1,len(df)):
    df['volatility'][i]=df['stock_price'][i]/df['stock_price'][i-1]
```
